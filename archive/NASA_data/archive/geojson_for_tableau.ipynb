```python
import os

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import geojsoncontour
```

```python
# levels to draw contour lines at
levels = [-70, -60, -50, -40, -30, -20, 10, 20, 30, 40]
```

The colors of the map does't really matter because Tableau does't recongize color information anyway.

```python
# read lon and lat coordinates. FILES ARE IN OUR GOOGLE DRIVE
lon = pd.read_csv('./processed_min/lon.csv', index_col=0)
lat = pd.read_csv('./processed_min/lat.csv', index_col=0)

# mesh x and y (lon and lat coordinates)
x_mesh, y_mesh = np.meshgrid(lon, lat)

# z_mesh
z_mesh = pd.read_csv("./processed_min/0101.csv", index_col=0)

# create the contour plot
contourf = plt.contourf(x_mesh, y_mesh, z_mesh, linestyles='None', levels=levels)
```

```python
# convert matplotlib contourf to geojson file
os.makedirs("./geojson_files", exist_ok=True)
geojsoncontour.contourf_to_geojson(contourf, geojson_filepath="./geojson_files/0101.geojson")
```

In Tableau, click "Spatial file" and open whatever geojson file you just saved

First drag "Geometery" to "Detail" and then drag "Fill" to "Color"

After this you can do whatever you want with the map
