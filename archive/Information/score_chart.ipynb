```python
import pandas as pd
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./data/some_degree_days.csv')
```

```python
# To evaluate weights and scores in the scoring equation.

# df extracted(made) from csv files showing HDD, CDD annually(365 days) for each location. 
# Each csv file cand be generated and downloadable from https://www.degreedays.net/#generate
# I picked some major cities and some minor locations that might represent extreme weathers.
# I used 15.5 degree C of boundary temperature for HDD and CDD

df
```

```python
# Weights of 0.005 for HDD, 0.004 for CDD and 0.5 for MPID(Must Plug-in Days). 
def calscore():
    EV_score = []
    for i in range(len(df)):
        HDD = df.iloc[i, 1]
        MPID = df.iloc[i, 2]
        CDD = df.iloc[i, 3]
        score = 100 - HDD*0.005 - MPID*0.5 - CDD*0.004
        EV_score.append(score)        
    df['EV_SCORE'] = EV_score
    added_df = df.sort_values(by=['EV_SCORE'], ascending=False)
    return added_df
```

```python
calscore()
```
